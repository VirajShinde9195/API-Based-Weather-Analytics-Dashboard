This cell fetches weather data from the Open-Meteo API for specified Indian cities (Pune, Mumbai, Delhi). It then processes the raw JSON response into pandas DataFrames, concatenates them, and saves the combined data to a CSV file named `weather_raw.csv`, which is then downloaded.

```python
import requests
import pandas as pd

CITIES = {
    "Pune":   {"lat": 18.52, "lon": 73.85},
    "Mumbai": {"lat": 19.08, "lon": 72.88},
    "Delhi":  {"lat": 28.61, "lon": 77.21},
}

all_data = []

for city, coords in CITIES.items():
    url = (
        f"https://api.open-meteo.com/v1/forecast"
        f"?latitude={coords['lat']}&longitude={coords['lon']}"
        f"&hourly=temperature_2m,relative_humidity_2m,windspeed_10m,"
        f"precipitation,cloudcover,weathercode"
        f"&timezone=Asia/Kolkata"
    )
    response = requests.get(url, timeout=10)
    raw = response.json()["hourly"]

    df = pd.DataFrame(raw)
    df["city"] = city
    all_data.append(df)
    print(f"{city} done — {len(df)} rows")

final = pd.concat(all_data, ignore_index=True)
final.to_csv("weather_raw.csv", index=False)
print(f"\nTotal rows: {len(final)}")
print("Saved: weather_raw.csv")

from google.colab import files
files.download("weather_raw.csv")
```

```python
# Reload raw CSV
df = pd.read_csv("weather_raw.csv")

# Time fix
df["time"] = pd.to_datetime(df["time"])

# Features add
df["date"]        = df["time"].dt.date
df["hour"]        = df["time"].dt.hour
df["day_name"]    = df["time"].dt.day_name()

def get_period(h):
    if 5  <= h < 12: return "Morning"
    if 12 <= h < 17: return "Afternoon"
    if 17 <= h < 21: return "Evening"
    return "Night"

df["period"] = df["hour"].apply(get_period)

weather_map = {
    0: "Clear", 1: "Mainly Clear", 2: "Partly Cloudy",
    3: "Overcast", 45: "Fog", 48: "Fog",
    51: "Drizzle", 61: "Rain", 80: "Showers"
}
df["weather_label"] = df["weathercode"].map(weather_map).fillna("Other")

# Confirm
print("Shape:", df.shape)
print("Cities:", df["city"].value_counts().to_dict())
```

This cell performs univariate analysis by plotting histograms for key weather variables: `temperature_2m`, `relative_humidity_2m`, `windspeed_10m`, and `cloudcover`. Each histogram includes a vertical line indicating the mean of the respective variable, and the plots are saved as `step5_univariate.png`.

```python
cols = ["temperature_2m", "relative_humidity_2m", "windspeed_10m",
         "cloudcover"]
colors = ["#E8593C", "#185FA5", "#1D9E75", "#E8A838"]

fig, axes = plt.subplots(1, len(cols), figsize=(18, 4))

for ax, col, color in zip(axes, cols, colors):
    ax.hist(df[col], bins=30, color=color, edgecolor="white")
    ax.axvline(df[col].mean(), color="black", linestyle="--", linewidth=1.5,
               label=f"Mean: {df[col].mean():.1f}")
    ax.set_title(col)
    ax.legend(fontsize=9)

plt.suptitle("Univariate Analysis — all variables", fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig("step5_univariate.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

cols = ["temperature_2m", "relative_humidity_2m", "windspeed_10m",
         "cloudcover"]
COLORS = {"Pune": "#E8593C", "Mumbai": "#185FA5", "Delhi": "#1D9E75"}

fig, axes = plt.subplots(1, len(cols), figsize=(18, 5))

for ax, col in zip(axes, cols):
    data = [df[df["city"] == city][col].values for city in COLORS]
    bp = ax.boxplot(data, patch_artist=True,
                    medianprops=dict(color="white", linewidth=2))
    for patch, color in zip(bp["boxes"], COLORS.values()):
        patch.set_facecolor(color)
        patch.set_alpha(0.8)
    ax.set_xticklabels(COLORS.keys(), rotation=15)
    ax.set_title(col)

plt.suptitle("City-wise Bivariate Analysis — Box Plots", fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig("step6_citywise.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

COLORS = {"Pune": "#E8593C", "Mumbai": "#185FA5", "Delhi": "#1D9E75"}

fig, ax = plt.subplots(figsize=(13, 5))

for city, grp in df.groupby("city"):
    # Hourly
    ax.plot(grp["time"], grp["temperature_2m"],
            color=COLORS[city], alpha=0.2, linewidth=0.8)

    # Daily average
    daily = grp.set_index("time")["temperature_2m"].resample("D").mean()
    ax.plot(daily.index, daily.values,
            color=COLORS[city], linewidth=2.5, label=f"{city} (daily avg)")

ax.set_title("Temperature Time Series — Hourly + Daily Average", fontsize=13)
ax.set_xlabel("Date")
ax.set_ylabel("Temperature (°C)")
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"))
ax.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("step7_timeseries.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot = df.pivot_table(
    values="temperature_2m",
    index="city",
    columns="hour",
    aggfunc="mean"
).round(1)

fig, ax = plt.subplots(figsize=(16, 3))
sns.heatmap(pivot,
            cmap="RdYlBu_r",
            annot=True,
            fmt=".0f",
            linewidths=0.3,
            cbar_kws={"label": "°C"},
            ax=ax)
ax.set_title("Average Temperature — City × Hour of Day", fontsize=13)
ax.set_xlabel("Hour of Day")
ax.set_ylabel("")
plt.tight_layout()
plt.savefig("step8_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nPeak hour per city:")
print(pivot.idxmax(axis=1))
```
